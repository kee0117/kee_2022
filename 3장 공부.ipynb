```python
numbers = [1, 2, 3, 4, 5]
result = []
for n in numbers:
    if n % 2 == 1:
        result.append(n*2)
    print(result)
```

```python
number = 0
number = int(input())

if number >= 10:
    print("구구단은 9단까지 입니다.")
elif number == 0:
    print("종료합니다")
else:
    print("%d단을 시작 합니다." % number)
    for num in range(1,10):
        print("%d * %d = " % (number,num) ,number * num)
```

```python
number = 0
number = int(input("숫자를 입력하세요: "))

if number in range(1,10):
    print("%d단을 시작 합니다." % number)
    for num in range(1,10):
        print("%d * %d = " % (number,num) ,number * num)
elif number == 0:
    print("종료합니다")
else:
    print("정수 1~9중에서 입력 하세요.")      
```
